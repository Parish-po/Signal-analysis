```python
import sys
sys.path.append("../src")


from preprocessing import load_dreamer_data        

df = load_dreamer_data("../data/raw/DREAMER.mat")
df.head()


```

```python
import matplotlib.pyplot as plt
import numpy as np

sig_raw = np.array(df.loc[0, "stimuli"])
plt.plot(sig_raw[:2000, 0])
plt.title("Filtered EEG (first channel, first video)")
plt.xlabel("Samples (time)")
plt.ylabel("Voltage (µV)")
plt.show()
```

```python
from scipy.fft import rfft, rfftfreq
import matplotlib.pyplot as plt
import numpy as np

signal = np.array(df.loc[0, "stimuli"])[:,0]   # first channel
fs = 128
N = len(signal)
freqs = rfftfreq(N, 1/fs)
fft_mag = np.abs(rfft(signal))

plt.semilogy(freqs, fft_mag)
plt.xlim(0, 60)
plt.title("EEG Power Spectrum after Filtering")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.show()
```
